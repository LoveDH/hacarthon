# 2. 데이터 전처리

## 1) 고객정보

- `개인사업자여부(PSN_BIZR_YN)`: 'Y'를 제외하고 결측치로 남아있습니다. 'N'으로 채웁니다.
- `생년월(TYMD)`: 생년(YEAR)과 생월(MONTH)로 분리하겠습니다. 타입은 int로 변환합니다.
- `주소`: 행정동명(CUS_ADMB_NM)은 결측치가 많아 삭제하고, 행정시도명(CUS_ADM_TRY_NM)과 시군구명(CUS_N_ADMN_NM)는 하나로 합칩니다.

```python
# 개인 사업자 여부
customer['PSN_BIZR_YN'].fillna('N',inplace=True)

# 생년월
customer['TYMD'] = customer['TYMD'].astype(str)
customer['YEAR'] = customer['TYMD'].str.slice(stop=4).astype(int)
customer['MONTH'] = customer['TYMD'].str.slice(start=4,stop=6).astype(int)
customer.drop('TYMD', axis=1, inplace=True)

# 주소

customer.update(customer[['CUS_ADM_TRY_NM','CUS_N_ADMZ_NM']].fillna(''))
customer.loc[customer['CUS_ADM_TRY_NM'].str.startswith('세종'), 'CUS_N_ADMZ_NM'] = '' # 세종시의 시군구명은 세종시 출범 전 구역도 있어 혼란방지를 위해 삭제
customer['ADDRESS'] = customer['CUS_ADM_TRY_NM'] + ' ' + customer['CUS_N_ADMZ_NM']
customer['ADDRESS'] = customer['ADDRESS'].str.strip() # 공백제거
customer.drop(['CUS_ADM_TRY_NM','CUS_N_ADMZ_NM', 'CUS_ADMB_NM'],axis=1, inplace=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

cols = ['PSN_BIZR_YN','SEX_SCN_NM','YEAR','MONTH','ADDRESS', 'CLB_HOUS_PYG_NM','REAI_BZTC_AVG_PCE']
train_cols = ['PSN_BIZR_YN','SEX_SCN_NM','YEAR','MONTH','ADDRESS']
customer_train = customer[cols]

le = LabelEncoder()
for col in train_cols:
    customer_train[col] = le.fit_transform(customer_train[col])
    
result = []
for to_predict in ['CLB_HOUS_PYG_NM','REAI_BZTC_AVG_PCE']:
    
    train, test = customer_train[customer_train[to_predict].notna()], customer_train[customer_train[to_predict].isna()]
        
    X_train, y_train, X_test = train[train_cols], train[to_predict], test[train_cols]
    dt_clf = DecisionTreeClassifier()
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.pred(X_test)
    
    test[to_predict] = pred
    test[to_predict] = test[to_predict].round(0)
    
    result.append(pd.concat([train,test], axis=0))
```

```python
customer['CLB_HOUS_PYG_NM'] = customer['CLB_HOUS_PYG_NM'].str.extract('(\d+)').astype(float)
customer['CLB_HOUS_PYG_NM'].fillna((customer['CLB_HOUS_PYG_NM'].mean() + customer['CLB_HOUS_PYG_NM'].median())/2, inplace=True)
customer['REAI_BZTC_AVG_PCE'].fillna((customer['REAI_BZTC_AVG_PCE'].mean() + customer['REAI_BZTC_AVG_PCE'].median())/2, inplace=True)
```
